# SciKit learn Multi Layer Perceptron (MLP) examples

Using MLP regressor to describe the connection between two variables. How to setup, train and apply an MLP regressor

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

#create 1000 numbers from 0-100 in steps of 0.1, and then change it into a [1000x1] array, for using in nn.fit()
x  = np.arange(0.0,100,0.01).reshape(-1,1)
#ravel() makes sure that y is a [1000x1] array and not a vector of 1000 values
y  = np.square(x).ravel()

# plot the result
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(x, y, s=5, c='g', marker="s", label='Data y = x^2')
ax1.legend(loc="lower right")
ax1.set_ylim(0.1,12000)
ax1.set_yscale("linear")
```

```python

# create the multi-layer perceptron (MLP) model with some parameters
nn = MLPRegressor(hidden_layer_sizes=(20,20),  activation='relu',
                solver='adam', max_iter=1000)

# train the model
print("Starting fit...")
n = nn.fit(x, y)
print("...done")

# Predict values using the MLP regressor
test_x = np.arange(0.0, 500, 1).reshape(-1,1)
test_y = nn.predict(test_x)

# plot the result of the fit
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(x, y, s=5, c='g', marker="s", label='Data y = x^2')
ax1.legend(loc="lower right")
ax1.set_xlim(0.1,100)
ax1.set_ylim(0.1,12000)

ax1.set_yscale("linear")

ax2.scatter(x, y, s=5, c='g', marker="s", label='Data')
ax2.plot(test_x,test_y, "r-", label='NN Prediction')
ax2.legend(loc="lower right")
ax2.set_ylim(0.1,12000)
ax2.set_xlim(0,100)
ax2.set_yscale("linear")
plt.tight_layout()
plt.show()
```

# MLP classifier example

Use same Iris dataset as for AdaBoostClassifier

```python
# sklearn comes with some example data sets
from sklearn import datasets

# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
#Import scikit-learn MLP classifier
from sklearn.neural_network import MLPClassifier 

# load the "iris" data set
iris = datasets.load_iris()

# get feature and target data for each object
X = iris.data
print(X)
y = iris.target

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test

# Create MLP classifer object
mlp = MLPClassifier(solver='adam', hidden_layer_sizes=(15, 5), max_iter=10000)

# Train MLP Classifer
model = mlp.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print(y_test)
print(y_pred)

# Confusion matric, which objects are misclassified? 
print(metrics.confusion_matrix(y_test, y_pred))
```

# Let's look at some other examples - how does training of MLP change with data size and number of iterations

```python
# synthetic classification dataset
from sklearn.datasets import make_classification
# define dataset
X, y = make_classification(n_samples=100000, n_features=10, n_informative=5, n_redundant=5, flip_y = 0.05, class_sep = 0.1, random_state=1)

# 10000 objects with 10 features each, 10000 target values
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

# split into training and testing sample
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

# 7000 training objects, 3000 test objects
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#Import scikit-learn MLP classifier
from sklearn.neural_network import MLPClassifier
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

trials = 40

acc_train = np.empty(trials)
acc_test = np.empty(trials)
iter = np.empty(trials)
iter[0] = int(10)

for i in range(trials):
   
    print(f"Trial {i}, max_iter = {iter[i]}")
    mlp = MLPClassifier(solver='adam', hidden_layer_sizes=(20, 20), max_iter=int(iter[i]))
    print(f"train model for {int(iter[i])} iterations")
    mlp.fit(X_train,y_train)
    print("done")
    
    y_pred1 = mlp.predict(X_train)
    y_pred2 = mlp.predict(X_test)
    acc_train[i] = metrics.accuracy_score(y_train, y_pred1)
    acc_test[i] = metrics.accuracy_score(y_test, y_pred2)

    print("Accuracy on training data:",acc_train[i])
    print("Accuracy on test data:",acc_test[i])
    print()
    
    if(i < trials-1):
        iter[i+1] = iter[i] * 1.2


```

```python
f,ax = plt.subplots(figsize = (12,4))
plt.plot(iter,acc_train,c="blue",label= "Training sample")
plt.plot(iter, acc_test,c="red",label = "Test sample")
plt.ylim(0.5,1.0)
plt.xlabel("No of iterations")
plt.ylabel("Model accuracy")
plt.xscale("log")
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, flip_y = 0.05, class_sep = 0.1, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

trials = 20

acc_train = np.empty(trials)
acc_test = np.empty(trials)
depth = np.empty(trials)

for i in range(trials):
    depth[i] = i+1
    print(f"Trial {i}, max_depth = {depth[i]}")
    clf = DecisionTreeClassifier(max_depth = int(depth[i]))
   
    print(f"train model for {int(iter[i])} iterations")
    clf.fit(X_train,y_train)
    print("done")
    
    y_pred1 = clf.predict(X_train)
    y_pred2 = clf.predict(X_test)
    acc_train[i] = metrics.accuracy_score(y_train, y_pred1)
    acc_test[i] = metrics.accuracy_score(y_test, y_pred2)

    print("Accuracy on training data:",acc_train[i])
    print("Accuracy on test data:",acc_test[i])
    print()
    
```

```python
f,ax = plt.subplots(figsize = (12,4))
plt.plot(depth,acc_train,c="blue",label= "Training sample")
plt.plot(depth, acc_test,c="red",label = "Test sample")
plt.ylim(0.5,1.0)
plt.xlabel("Depth of decision tree")
plt.ylabel("Model accuracy")
plt.legend()
plt.show()
```
