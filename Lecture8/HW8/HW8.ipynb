# Q4:

**K-means 聚类算法**：无监督算法

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import v_measure_score

kmeans = KMeans(n_clusters=3).fit(df2[["x0", "x1", "x3"]])
y_pred = kmeans.predict(df2[["x0", "x1", "x3"]])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

axes[0].scatter(df2["x0"], df2["x1"], c=df2["y"])
axes[0].set_title("Input clusters")

axes[1].scatter(df2["x0"], df2["x1"], c=y_pred)
axes[1].set_title("Predicted clusters")

v_measure = v_measure_score(df2["y"], y_pred)

# 打印 v_measure_score
print(f"v_measure_score: {v_measure}")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import v_measure_score

# 读取 CSV 文件
df3 = pd.read_csv("data_kmeans_HW8.txt")

# 创建 KMeans 对象并进行训练
kmeans = KMeans(n_clusters=3)
kmeans.fit(df3[["x0", "x1", "x2"]])

# 获取预测结果
y_pred = kmeans.labels_

# 计算 v_measure_score
v_measure = v_measure_score(df3["y"], y_pred)

# 打印 v_measure_score
print(f"v_measure_score: {v_measure}")
```
