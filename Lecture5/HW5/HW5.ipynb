```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data_HW5.csv")

# 绘制 x1 的直方图，并根据类别进行颜色区分
sns.histplot(data=df, x="x1", hue="Type", element="step")
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.show()

# 绘制 x1 的核密度估计图，并根据类别进行颜色区分
sns.kdeplot(data=df, x="x1", hue="Type")
plt.xlabel("x1")
plt.ylabel("Density")
plt.show()
```
