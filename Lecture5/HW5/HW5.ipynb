```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data_HW5.csv")

#  x1 
sns.histplot(data=df, x="x1", hue="Type", element="step")
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.show()

#  x1 
sns.kdeplot(data=df, x="x1", hue="Type")
plt.xlabel("x1")
plt.ylabel("Density")
plt.show()
```
